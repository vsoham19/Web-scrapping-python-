```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.meesho.com/speakers/pl/9ub"
```

```python
response = requests.get(url)
html_content = response.text
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
print(soup.prettify())

```

```python
divs = soup.find_all('div')
```

```python
for div in divs:
    classes = div.get('class')  
    if classes:  
        print(classes)
```

```python
product_names = [item.text for item in soup.find_all(class_='NewProductCardstyled__ProductHeaderWrapper-sc-6y2tys-32')]
product_prices = [item.text for item in soup.find_all(class_='NewProductCardstyled__PriceRow-sc-6y2tys-7')]
```

```python
data = {
    'Product Name': product_names,
    'Price': product_prices
}
df = pd.DataFrame(data)
```

```python
df.to_csv('speakers.csv', index=False)
```
